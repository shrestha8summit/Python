```python
name = []    #forming an empty list for input
roll = []
age = []
address = []
phone = []
gpa = []
r_name = [] #forming empty list for storing removed items 
r_roll = []
r_age = []
r_address = []
r_phone = []
r_gpa = []
def num():
    print("Please enter a valid number.")
def correction():
    try:
        print("---------------------------------------------------------------------------------")
        print("S.N\t NAME\t ROLL\t AGE \t ADDRESS\t PHONE\t\tGPA")
        for i in range(len(o_name)):
            print("---------------------------------------------------------------------------------")
            print(i + 1, "\t", o_name[i], "\t", o_roll[i], "\t", o_age[i], "\t", o_address[i], "\t", o_phone[i],
                  "\t", o_gpa[i])
            print("---------------------------------------------------------------------------------")
    except Exception as e:
        print(f"An error occurred while displaying data: {e}")
def re():
    try:
        sn = int(input("Enter the s.n whose data is to be corrected => "))
        sn = sn - 1
        oop = int(input("What do you want to edit\n1.Name\n2.Roll\n3.Age\n4.Address\n5.Phone\n6.GPA => "))
        if oop == 1:
            o_name[sn] = input("\nEnter the correct name: ")
        elif oop == 2:
            o_roll[sn] = int(input("\nEnter the correct roll: "))
        elif oop == 3:
            o_age[sn] = int(input("\nEnter the correct age: "))
        elif oop == 4:
            o_address[sn] = input("\nEnter the correct address: ")
        elif oop == 5:
            o_phone[sn] = int(input("\nEnter the correct phone number: "))
        else:
            o_gpa[sn] = float(input("\nEnter the correct GPA: "))
    except (ValueError, IndexError) as e:
        print(f"An error occurred while editing data: {e}")
# Taking input from the user
print("\t\t Entry for new records of the latest academic year")
n = 1
while n == 1:
    namee = input("Enter your name: ")
    rolll = int(input("Enter your roll number: "))
    while True:
        try:
            ageee = int(input("Enter your age:\t"))
            assert ageee >= 0, "Age must be a positive integer."
            agee=ageee
            #print("So, your age is:", ageee)
        except ValueError:
            print("Invalid input. Please enter a valid integer for age.")
        except AssertionError as e:
            print("Assertion Error:", str(e))
        if ageee>0:
            break
    addresss = input("Enter your home address: ")
    while True:
        try:
            phonee = int(input("Enter your phone number: "))
            pho = str(phonee)
            if len(pho) != 10:
                num()
            else:
                break
        except ValueError:
            num()
    m1 = float(input("Enter the marks obtained in physics: "))
    m2 = float(input("Enter the marks obtained in maths: "))
    m3 = float(input("Enter the marks obtained in N.C: "))
    m4 = float(input("Enter the marks obtained in PTTL: "))
    m5 = float(input("Enter the marks obtained in c prog: "))
    per = (m1 + m2 + m3 + m4 + m5) / 500
    perc = per * 100
    gpaa = perc / 9.5
    name.append(namee)
    roll.append(rolll)
    age.append(agee)
    address.append(addresss)
    phone.append(phonee)
    gpa.append(gpaa)
    n = int(input("\nIs there any data remaining for entry? 1 for yes, 0 for no: "))
    print("\n")
# Making a list from past year academic record
o_name = ["Rajan", "Sujal", "Abin", "Ayush", "Sagar"]
o_roll = [1, 2, 3, 4, 5]
o_age = [18, 19, 19, 17, 20]
o_address = ["Kathmandu", "Lalitpur", "Kirtipur", "Kalikot", "Nepalgunj"]
o_phone = [9847362467, 9847362468, 9847362469, 9846354983, 9856084329]
o_gpa = [8.9, 9, 8.7, 5, 7.5]
# List operations
o_name.extend(name)
o_roll.extend(roll)
o_age.extend(age)
o_address.extend(address)
o_phone.extend(phone)
o_gpa.extend(gpa)
# Printing the list of records
print("---------------------------------------------------------------------------------")
print("S.N\t NAME\t ROLL\t AGE \t ADDRESS\t PHONE\t\tGPA")
for i in range(len(o_name)):
    print("---------------------------------------------------------------------------------")
    print(i + 1, "\t", o_name[i], "\t", o_roll[i], "\t", o_age[i], "\t", o_address[i], "\t", o_phone[i], "\t", o_gpa[i])
print("---------------------------------------------------------------------------------")
# Check if any past year students left the college
drop = int(input("\nDid any of the past year students leave our college? 1 for yes, 0 for no: "))
if drop == 1:
    while drop == 1:
        for i in range(0, len(o_name)):
            try:
                which = int(input("Enter the S.N from the last list of students who have left college: ")) - 1
                re_name = o_name.pop(which)
                re_roll = o_roll.pop(which)
                re_age = o_age.pop(which)
                re_address = o_address.pop(which)
                re_phone = o_phone.pop(which)
                re_gpa = o_gpa.pop(which)
                r_name.append(re_name)
                r_roll.append(re_roll)
                r_age.append(re_age)
                r_address.append(re_address)
                r_phone.append(re_phone)
                r_gpa.append(re_gpa)
                again = int(input("Is there any other student who left college? 1 for yes, 0 for no: "))
                if again == 1:
                    drop = 1
                else:
                    break
            except (ValueError, IndexError) as e:
                print(f"An error occurred while removing a student: {e}")
    print("\nThe list of students who have left our college:\n")
    print("---------------------------------------------------------------------------------")
    print("S.N\t NAME\t ROLL\t AGE \t ADDRESS\t PHONE\t\tGPA")
    for i in range(len(r_name)):
        print("---------------------------------------------------------------------------------")
        print(i + 1, "\t", r_name[i], "\t", r_roll[i], "\t", r_age[i], "\t", r_address[i], "\t", r_phone[i], "\t", r_gpa[i])
        print("---------------------------------------------------------------------------------")
else:
    print("No students from the past year left the college.")
# Calculation in the management file for the operations involved in list, tuple, and dictionary
t_o_name = tuple(o_name)
t_o_roll = tuple(o_roll)
t_o_age = tuple(o_age)
t_o_address = tuple(o_address)
t_o_phone = tuple(o_phone)
t_o_gpa = tuple(o_gpa)
edit = int(input("\nDo you want to edit any data? 1 for yes, 0 for no: "))  # List manipulation
print("Original record of students for editing")
if edit == 1:
    correction()
    count = int(input("How many data points are you going to update: "))
    for i in range(0, count):
        re()
# Printing the result of the class
print("\nHere is the list of results of students from the class:")
print("\n---------------------------------------------------------------------------------")
record = []
record = o_gpa.copy()
record.sort(reverse=True)
print("Rank\t NAME\t ROLL\t AGE \t ADDRESS\t PHONE\t\tGPA")
print("---------------------------------------------------------------------------------")
for i in range(0, len(o_name)):
    recordd = record[i]
    position = o_gpa.index(recordd)
    print(i + 1, "\t", o_name[position], "\t", o_roll[position], "\t", o_age[position], "\t", o_address[position], "\t", o_phone[position], "\t", o_gpa[position])
    print("---------------------------------------------------------------------------------")
top = int(input("\nDo you want to know who achieved the highest GPA among all the students? 1 for yes, 0 for no: "))
if top == 1:
    print("The highest GPA is:", max(t_o_gpa))
    dherai = o_gpa.index(max(o_gpa))
    print("\nThe details of the person are given below:")
    print("---------------------------------------------------------------------------------")
    print(o_name[dherai], "\t", o_roll[dherai], "\t", o_age[dherai], "\t", o_address[dherai], "\t", o_phone[dherai], "\t", o_gpa[dherai])
    print("---------------------------------------------------------------------------------")
low = int(input("\nDo you also want to know who achieved the lowest GPA among all the students? 1 for yes, 0 for no: "))
if low == 1:
    print("The lowest GPA is:", min(t_o_gpa))
    thorai = o_gpa.index(min(o_gpa))
    print("\nThe details of the person are given below:")
    print("---------------------------------------------------------------------------------")
    print(o_name[thorai], "\t", o_roll[thorai], "\t", o_age[thorai], "\t", o_address[thorai], "\t", o_phone[thorai], "\t", o_gpa[thorai])
    print("---------------------------------------------------------------------------------")
    print("Hope the students will get even better education.")
```
